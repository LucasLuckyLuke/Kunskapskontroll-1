# Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

It can be done with the object oriented approach or in pyplot / MATLAB style. The recommended way in general is the object oriented approach, especially if it's for a complicated plot or plots, or if you for example are doing a big project that also is complicated. Pyplot / Matlab can be good in certain situation when doing quick interactive work. But for the most part object oriented is the way to go.

# Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
Figure can be seen as a canvas. Or for example a whiteboard. It's like a white wall for the figure to be on.
Axes is what's on the figure, can be imagined as what we put on the whiteboard or canvas, (figure). 
A figure can contain as many axes as we want. Axis sets the limitations of our plot that is put on a figure, or if
you wanna explain it better axis decides what scale and limits there gonna be but also decide the generate ticks.
Artist is everything that is on the figure. That's being drawn, for example the titles, lines, text, shapes.
```

# Create a figure containing 2 subplots where the first is a scatter plot and the second is a bar plot. You have the data below.

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.scatter(random_data_x, random_data_y, label='Random Data')
ax1.plot(x, y, color='red', label='x^2')
ax1.set_title('Scatter Plot')
ax1.set_xlabel('X Axis')
ax1.set_ylabel('Y Axis')

ax2.bar(names, values, color='green')
ax2.set_title('Bar Plot')
ax2.set_xlabel('Fruits')
ax2.set_ylabel('Quantity')

plt.tight_layout()

plt.show()
```
